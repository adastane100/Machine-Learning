We will call resample with the rule parameter, then you need to call some sort of aggregation function. The rule paramter will tell the basis of resampling.

```python
#yearly means
data.resample(rule='A').mean().head()
```

```python
data.resample(rule='A').max().head()
```

## Custom Resampling

```python
def custom_resampling(entry):
    return entry.mean()
```

```python
data.resample(rule='A').apply(custom_resampling).head()
```

# Rolling and Expanding

```python
data['Open'].plot(figsize=(16,6))
```

Since the data is noisy we have to average the data by the week.

```python
data.rolling(7).mean().head(20)
```

```python
data['Open'].plot()
data.rolling(window=30).mean()['Close'].plot(figsize=(16,6))
```

```python
data['Close 30 days Moving Avg'] = data['Close'].rolling(window=30).mean()
data[['Close 30 days Moving Avg','Close']].plot(figsize=(16,6))
```

## Expanding

```python
data['Close'].expanding().mean().plot(figsize=(16,6))
```

```python
data['Close after 20 days MA'] = data['Close'].rolling(window=20).mean()
data['Upper'] = data['Close after 20 days MA'] + 2*data['Close'].std()
data['Lower'] = data['Close after 20 days MA'] - 2*data['Close'].std()
data[['Close after 20 days MA','Close','Upper','Lower']].plot(figsize=(16,6))
```

# Exploring ETS,EWMA Models on the stock

```python
df = quandl.get('NSE/HINDUNILVR')
```

```python
df.head()
```

```python
df['SMA-6months'] = df['Close'].rolling(window=180).mean()
df['SMA-1YEAR'] = df['Close'].rolling(window=365).mean()
```

```python
df[['Close','SMA-6months','SMA-1YEAR']].plot(figsize=(16,6))
```

```python
df['EWMA-1YEAR Close'] = df['Close'].ewm(span=365,adjust=True).mean()
```

```python
df[['Close','EWMA-1YEAR Close']].plot(figsize=(10,8))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df['Close'],model='multiplicative',freq=52)
```

```python
fig = result.plot()
fig.set_size_inches(15, 8)
```
